```python
def compute_simple_payout(df, attr_model, ascending=False, bet_amount=1.0):
    """
    Add columns for quick calculation of Win bets % payout
    :param df: Dataframe from dataset for multiple races
    :param attr_model: (string)an attribute / prob or score in the dataframe that can be ranked
    :param ascending: False if best is higher score i.e. probs
    :param bet_amount: default to 1.0
    :return: Dataframe with columns added

    """
    df['is_win'] = df['official_finish_position'].map(lambda x:int(x==1))
    df['rank_' + attr_model] = df.groupby('race_id')[attr_model].transform(lambda x: x.rank(ascending=False))
    df['bet_amount'] = df['rank_' + attr_model].map(lambda x:int(x<1.5)*bet_amount)
    df['is_wager'] = df['bet_amount'].map(lambda x:int(x>0))
    df['is_paid'] = df['is_wager'] * df['is_win']
    df['payout'] = df['is_win']*df['bet_amount']*df['payout_win'].fillna(0.0)

    return df
```

```python
def compute_advantage(df):
    try:
        #print(df.race_id).iloc[0]
        pct_win = df.groupby('race_id')['is_paid'].sum().value_counts(normalize=True)[1]
        
    except:
        
        return 0.0

    pct_loss = 1.0 - pct_win
    mean_odds = df[df.is_paid>0]['final_tote_odds'].mean()
    advantage = pct_win - pct_loss / mean_odds
    #print(advantage)
    return advantage
```

```python
#HERE WE ADD columns which are based on the runners withing a race and some aggregating function on the scores
#i.e. we would like to kno how many runners have 'x8is_HDWPSRRating_norm_par' ==1 , which is the number of runners who 
#are prjecte to run as fast as the averag winning speed for this race type an track



def add_agg_factor_column(df, grpby_columns, factor_column, aggfunc, prefix):
    #Adds a column to df by applying an agg function to the factor by group
    df["x8agg"+"_"+factor_column+prefix] = df.groupby(grpby_columns)[factor_column].transform(aggfunc)
    print("added column {}".format("x8agg"+"_"+factor_column+"_"+prefix))
    return df

#add column based on x8is_HDWPSRRating_norm_par number of runners fast enough in each race
df_factors_PILOT = add_agg_factor_column(df_factors_PILOT, grpby_columns='race_id',factor_column='x8is_HDWPSRRating_norm_par', aggfunc=sum, prefix='sum')
```

```python
#THESE ARE TO ADD transforms based on the previous performances for each runner

def range_score_pp(df, past_performance_columns, target_factor_column, max_range=None): #if max_range is None allows for nmbers higher than 1
    #computes the target relative to range normalized
    max_pp = df[past_performance_columns].max(axis=1)
    min_pp = df[past_performance_columns].min(axis=1)
    current = df[target_factor_column]
    return (current - min_pp)/(max_pp - min_pp)

def add_pp_factor_column(df, input_columns, target_column, seqfunc, prefix):
    #Adds a column based on a sequence of _pp values in a row
    df[prefix] = seqfunc(df, input_columns, target_column)
    print("added {}".format(prefix))
    return df

    

df_factors_PILOT = add_pp_factor_column(df_factors_PILOT, input_columns = [col for col in df_factors_PILOT.columns if col.startswith('HDWSpeedRating')], target_column='HDWPSRRating', seqfunc=range_score_pp, prefix='rsi_HDWPSRRating')
```

```python
# create a dictionary of advantages. Keys are factor names
def compute_advantage_by_group(df, grouping_columns, _factors):
    """
    groups by grouping columns and computes advantages
    """
    grouped = df.groupby(grouping_columns)
    grp_dict = {}
    for grp, dfgrp in grouped:
        advantage = {}
        for factor in _factors:
            df = dfgrp.copy()
            attr_bench_final = factor
            df_new = compute_simple_payout(df, attr_model=attr_bench_final, ascending=False, bet_amount=1.0).copy()
            advantage[factor] = compute_advantage(df_new)
        grp_dict[grp] = advantage
    dfgrp = pd.DataFrame(grp_dict)
    dfgrp.columns.names = grouping_columns
    return dfgrp
```
