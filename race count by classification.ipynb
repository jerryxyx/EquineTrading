```python
import pandas as pd
import numpy as np
dfres = pd.read_csv('I:/YaoTony/dfres.csv', low_memory=False) 
```

```python
# First make race unique
dfres = dfres.drop_duplicates(subset=['race_id'], keep='last')
```

```python
dfres = dfres[['race_type_code', 'sex_restriction_code', 'age_restriction_code', 'distance', 'x8_track_sym']]
```

```python
reshaped_dfres = dfres.pivot_table(index=['race_type_code',
                                          'sex_restriction_code',
                                          'age_restriction_code',
                                          'distance'],
                                   columns='x8_track_sym',
                                   aggfunc=len, 
                                   fill_value=0).reset_index()
reshaped_dfres
```

```python
reshaped_dfres.to_csv('I:/YaoTony/reshaped_dfres.csv', encoding='utf-8')
```
