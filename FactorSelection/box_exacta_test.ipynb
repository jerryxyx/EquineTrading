```python
df_CALIBRATE = pd.read_csv('DF_CALIBRATE_20180629.csv')
df_CALIBRATE.head()
```

```python
df = df_CALIBRATE[['race_id','EX_payout','runner_program_number_pp','x8is_exacta','prob_x8runner_HDWPSRRating_norm_par']]
df.head()
```

```python
import numpy as np
df = df[np.isfinite(df['prob_x8runner_HDWPSRRating_norm_par'])]#delete rows when race only has one runner
df['is_paid']=0
race_set = set(df['race_id'])
for race in race_set:
    df_sub = df[df['race_id']==race]
    #fav is prob ranking 1 of box combination and its prob
    fav = Harville.harville_box(list(df_sub['runner_program_number_pp']),list(df_sub['prob_x8runner_HDWPSRRating_norm_par']),'Exacta')[0]
    df_sub_ex = df_sub[df_sub['x8is_exacta']>0]#extract extacta runners
    if set(df_sub_ex['runner_program_number_pp'])==set(fav[0:2]):
        df.loc[df['race_id']==race,'is_paid']=1
```
